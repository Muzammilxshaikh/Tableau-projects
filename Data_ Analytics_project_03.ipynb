## specifying new column names that i want to use

bikes.columns

```python
new_column_dict = {
    "timestamp":"time",
    "cnt":"count",
    "t1":"temp_real_C",
    "t2":"temp_feels_like_C",
    "hum":"humidity_percent",
    "wind_speed":"wind_speed_kmp",
    "weather_code":"weather",
    "is_holiday":"Is_holiday",
    "is_weekend":"is_weekend",
    "season":"season"
}

#rename columnname

bikes.rename(new_column_dict, axis=True, inplace=True)
```

```python
bikes.sample(10)
```

```python
bikes.humidity_percent=bikes.humidity_percent/100
```

```python
bikes.humidity_percent
```

```python
season_dict = {
    "0.0":"spring",
    "1.0":"summer",
    "2.0":"autumn",
    "3.0":"winter"
}
```

```python
weather_dict = {
    "1.0":"Clear",
    "2.0":"Scattered clouds",
    "3.0":"Broken clouds",
    "4.0":"Cloudy",
    "7.0":"Rain",
    "10.0":"Rain with thunderstorm",
    "26.0":"Snowfall"
}
```

```python
## changing the data type

print(bikes['season'].dtype)
```

```python
bikes.season=bikes.season.astype('str')
```

```python
## changing the values with dictionary

bikes.season = bikes.season.map(season_dict)
```

```python
print(bikes["weather"].dtype)
```

```python
bikes.weather = bikes.weather.astype('str')
```

```python
bikes.weather = bikes.weather.map(weather_dict)
```
